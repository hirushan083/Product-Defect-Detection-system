```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Image size & batch size
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Path to test dataset folder
test_dir  = r"C:\Users\User\OneDrive\Desktop\data\defect_detector\casting_data\casting_data\test"

test_datagen = ImageDataGenerator(rescale=1./255)

# Load test data
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size = IMG_SIZE,
    batch_size = BATCH_SIZE,
    class_mode = 'binary',
    shuffle = False
)
```

```python
from tensorflow.keras.models import load_model

# Load the saved model
loaded_model = load_model("defect_detection_cnn_model.h5")
```

```python
predictions = loaded_model.predict(test_data)

# Convert probability values into class labels
predicted_classes = (predictions > 0.5).astype("int32") 

print(predicted_classes[:25])
```
